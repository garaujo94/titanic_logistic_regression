```python
#Getting the label in a different variable
y = df['Survived'].copy()
```

```python
#Creating a sparse matrix
X_train = pd.get_dummies(df.iloc[:,1:])
X_test = pd.get_dummies(df_test)
X_train.shape
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
#Dropping the column 'Sex_male' because it is explicit in the 'Sex_female'. If Sex_female is 1, so Sex_male is 0
X_train.drop(columns = 'Sex_male', inplace = True)
```

```python
X_test.drop(columns = 'Sex_male', inplace = True)
```

```python
#importing and fitting the model
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y)
```

```python
#Make predictions
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
#Getting labels
sub = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
y_test = sub.iloc[:,1].values
#Perfomance
print(classification_report(y_test, y_pred))
print('\n')
print(confusion_matrix(y_test, y_pred))
```

```python
my_submission = pd.DataFrame(y_pred)
my_submission.to_csv('submission.csv', index = False)
```
